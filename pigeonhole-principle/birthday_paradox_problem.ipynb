# Doğum günü problemi

Doğum günü problemi, rastgele seçilen bir grup insan için, bazı çiftlerin aynı doğum gününe sahip olma ihtimalinin ne olduğunu soruyor. Güvercin yuvası prensibine göre, eğer odada 367 kişi varsa, aynı doğum gününü paylaşan en az bir çiftin olduğunu biliyoruz, çünkü seçilebilecek yalnızca 366 doğum günü vardır (29 Şubat dahil).

```
from pylab import *
import numpy as np

# insan sayısı (n)
n = 367
# doğum günü sayısı (m)
m = 366

p = int(ceil(n/m))

print ("Senedeki en az bir gün", p, "insan içerir.")
```

# Doğum günü paradoksu

Doğum günü "paradoksu", grup 23 kişi kadar küçük olsa bile, aynı doğum gününe sahip bir çiftin olma ihtimalinin hala% 50'nin üzerinde olduğunu gösteriyor.

Bu problemin arkasındaki matematik olasılıklı modeli bir karma fonksiyonun çatlama karmaşıklığını azaltmak için kullanan **doğum günü paradoksu** adı verilen iyi bilinen bir şifreleme saldırısına yol açtı.

```
from pylab import *
import numpy as np
 
def makePlot(useYLogScale = False):
    N=91
    n = np.arange(float(N))
    pbar=np.exp(-n* (n-1) / (2.0*365.0))
    p=1.0-pbar
 
    n05 = 0.5*(1.0+np.sqrt(1-8.0*365.0*np.log(1.0-0.5)))
    plot([n05,n05],[0.0,0.5],c='k', linestyle='--')
    plot([0.0,n05],[0.5,0.5],c='k', linestyle='--')
    text(23.5,0.02,' ~23')

    plot(n,p   ,c='r',label = unicode('Çiftin olasılığı'))
    plot(n,pbar,c='b',label = unicode('Eşleşmeyen çiftin olasılığı'))
 
    legend(loc='right')
    xlim(0, N)
    if useYLogScale:
        ylim(1e-6, 1)
        ax = gca()
        ax.set_yscale('log')
    else:
        ylim(0, 1)
        yticks([0.0,0.2,0.4,0.5,0.6,0.8,1.0])
    xticks(range(0, N, 10))
    grid(True, ls='-', c='#a0a0a0')
    
    xlabel(unicode('İnsanların sayısı'))
    if useYLogScale:
        ylabel(unicode('Olasılık (log)'))
    else:
        ylabel(unicode('Olasılık'))

    show()
 
makePlot()
makePlot(useYLogScale = True)
```
